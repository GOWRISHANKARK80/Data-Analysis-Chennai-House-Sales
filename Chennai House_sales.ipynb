```python
val_col=["QS_ROOMS","QS_BATHROOM","QS_BEDROOM"]
target_col="QS_OVERALL"
def fill_null_QS_OVERALL(row):
  if pd.isnull(row[target_col]):
    return row[val_col].mean()
  else:
    return row[target_col]

df[target_col]=df.apply(fill_null_QS_OVERALL,axis=1)
df.loc[df["N_BATHROOM"].isnull(),"N_BATHROOM"]=df[df["N_BATHROOM"].isnull()].apply(lambda i: 2 if ((i["N_BEDROOM"] == 1) & (i["INT_SQFT"] >1800)) else 1,axis=1)

df.loc[df["N_BEDROOM"].isnull(),"N_BEDROOM"]=df[df["N_BEDROOM"].isnull()].apply(lambda i: 1 if ((i["N_BATHROOM"] == 1) & (i["INT_SQFT"] <1800)) else 2 ,axis=1)
```

```python
categories=['AREA','N_BEDROOM', 'N_BATHROOM','N_ROOM', 'SALE_COND', 'PARK_FACIL', 'BUILDTYPE', 'UTILITY_AVAIL','STREET', 'MZZONE','month_date_sale', 'day_date_sale', 'week_date_sale']
```

```python
continues=['INT_SQFT', 'DIST_MAINROAD', 'QS_ROOMS', 'QS_BATHROOM', 'QS_BEDROOM','QS_OVERALL', 'REG_FEE', 'COMMIS', 'SALES_PRICE', 'year_date_sale','build_age']
```

```python
#defining funtion for continues vs continues
from scipy import stats

def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.055:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False
```

```python
two_sample(df['INT_SQFT'],df['DIST_MAINROAD'])
```

```python
#defining function for categories vs categories
def chisqare_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.055 else False
```

```python
chisqare_cat_vs_cat(df['AREA'],df['N_BEDROOM'])
```

```python
#defining function for continous(d1) vs categories(d2)

def annova_test(d1,d2):
  group=df[d2].unique()
  data={}
  for i in group:
    data[i]=df[d1][df[d2]==i]
  f_value,p_value=stats.f_oneway(*[i for i in data.values()])
  if p_value < 0.055:
    return False
  else:
    return True
```

```python
annova_test('INT_SQFT','AREA')
```

```python
final={}
for i in df.columns:
  final[i]={}
  for j in df.columns:
    if (i in continues) and (j in continues):
      result=two_sample(df[i],df[j])
    elif  (i in continues) and (j in categories):
      result=annova_test(i,j)
    elif (i in categories) and (j in continues):
      result=annova_test(j,i)
    elif (i in categories) and (j in categories):
      result=chisqare_cat_vs_cat(df[i],df[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
```
